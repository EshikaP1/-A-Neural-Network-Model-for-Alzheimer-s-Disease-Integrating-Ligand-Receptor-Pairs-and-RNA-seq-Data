```python
# Step 1: Select resources for cell-cell communication
ccc_resources = liana.rs.select_resource("consensus")
ccc_resources.head()
```

```python
count=0
for x in ccc_resources['ligand']:
  placement_variable=[x,ccc_resources['receptor'].iloc[count]]
  print(placement_variable)
  count+=1
```

```python
combined_column = pd.concat([ccc_resources['ligand'], ccc_resources['receptor']], ignore_index=True).drop_duplicates()
print('There are {} unique genes in the LIANA consensus ligand-receptor database'.format(combined_column.shape[0]))
combined_column.head()
```

```python
import requests
import time

# Function to get gene symbol from Ensembl ID
#Filters the genes from human and mice; therefore, some of the genes will not have an Ensembel ID
def get_ensembl_id_by_gene_name(gene_name):
    url = f"https://rest.ensembl.org/xrefs/symbol/homo_sapiens/{gene_name}?"
    headers = {"Content-Type": "application/json"}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        data = response.json()
        for entry in data:
            if entry.get('type') == 'gene':
                return entry.get('id')  # Return the Ensembl gene ID
        return "Gene ID not found"
    else:
        return f"Error: {response.status_code}"
```

```python
geneId_list = []
geneName_list = []
```

```python
for geneName in combined_column:
  geneId_list.append(get_ensembl_id_by_gene_name(geneName))
  geneName_list.append(geneName)
```

```python
print(geneId_list)
print(geneName_list)
```

```python
df2 = pd.DataFrame({'geneName': geneName_list, 'geneId': geneId_list})
```
